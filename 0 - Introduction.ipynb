## If it's so simple, why all the fuss?

* Difficult to measure: these things move at around ~8 000 meters per second (Athens-Thessaloniki in one minute) and GPS precision is not that good
* ...But great accuracy is required!
* Many perturbations: the Earth is not a sphere, the Moon is very close, the sunlight pushes the satellite...
* If you lose contact with the satellite, it's a needle in a haystack

## Some (alternative) tools for Mission Analysis

* **AGI STK** https://www.agi.com/products (_commercial_)
  - GUI with scripting capabilities
  - Most powerful tool in the market
  - Also most expensive one, and closed source
* **Orekit** https://www.orekit.org/
  - Java library
  - Mature, widely used
  - Not so focused on interplanetary applications
* **SPICE** https://naif.jpl.nasa.gov/naif/toolkit.html
  - C, FORTRAN, IDL, MATLAB, Python... libraries
  - Battle tested collection of functions
  - Generating the kernels requires some expertise
* **GMAT** https://sourceforge.net/projects/gmat/
  - GUI with scripting capabilities
  - Closest open source competitor to STK
  - Sometimes unstable

* **poliastro** https://docs.poliastro.space/en/stable/
  - (Pure) Python library
  - Optimized for ease of use and contribution
  - Less powerful than the alternatives
